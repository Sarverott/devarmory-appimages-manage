```python
import json
import os
import pathlib

print(os.getcwd())

print(pathlib.Path(os.getcwd())/"..")

print((pathlib.Path(os.getcwd())/"..").resolve())

print(str((pathlib.Path(os.getcwd())/"..").resolve()))
```

```python
appsindex = None

with open((pathlib.Path(os.getcwd())/".."/"appsindex.json").resolve()) as appsindexfile:
    appsindex = json.load(appsindexfile)


print(json.dumps(appsindex['apps'],indent=4))
```

```python
appindex_path = (pathlib.Path(os.getcwd())/".."/"appsindex.json").resolve()
appitem_template_path = (pathlib.Path(os.getcwd())/".."/"resources"/"app.template.json").resolve()

appsindex = None

with open(appindex_path) as appsindexfile:
    appsindex = json.load(appsindexfile)

print(json.dumps(appsindex['apps'][0],indent=4))

with open(appitem_template_path, "w") as appitemfile:
    json.dump(appsindex['apps'][0],appitemfile)

```
